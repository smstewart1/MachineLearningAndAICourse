Start by importing the relevant libraries

```python
import pandas as pd
import keras
from sklearn.model_selection import train_test_split
from keras.models import Sequential
from keras.layers import Dense
```

Next load the datas a dataframe

```python
#read in the data
df = pd.read_csv("concrete_data.csv")

#break the data into the variates and the target 
dfc = df.columns
df_variates = df[dfc[dfc != 'Strength']]
df_target = df['Strength']

x_train, x_test, y_train, y_test = train_test_split(df_variates, df_target, test_size = 0.3)
```

Build a function

```python
def myNN(x_test, x_train, y_test, y_train):
    #build the model
    model = Sequential()
    n_cols = x_train.shape[1]
    model.add(Dense(10, activation = 'relu', input_shape = (n_cols, ))) #neutrons = # of neurons in the hidden layer, activation = function being used
    model.add(Dense(1)) #ouptu layer number of neurons = # of unique outcomes

    #optimize the model
    model.compile(optimizer = 'adam', loss = 'mean_squared_error') #defines our optimizer model, adam is a better gradient descent algorithm
    model.fit(x_train, y_train, epochs = 50)
    
    #test and validate the model
    scores = model.evaluate(x_test, y_test)
    return scores
```

Now we run the model

```python
results = []

for i in range(0, 50):
    throwback = myNN(x_test, x_train, y_test, y_train)
    results.append(throwback)

print(results)
```

Now computer the average RMSE

```python
avg = sum(results)/len(results)

print(avg)
```

Part B: same thing but with normalization

```python
#normalize the data
column_list = df_variates.columns

for i in column_list:
    df_variates[i] = (df_variates[i] - df_variates[i].mean())/df_variates[i].std()
```

```python
df_target = (df_target - df_target.mean())/df_target.std()
```

```python
#split the data
x_train, x_test, y_train, y_test = train_test_split(df_variates, df_target, test_size = 0.3)
```

Does the optimizatin again

```python
results = []

for i in range(0, 50):
    throwback = myNN(x_test, x_train, y_test, y_train)
    results.append(throwback)

print(results)
```

Calculate the new RMSE

```python
avg = sum(results)/len(results)

print(avg)
```
